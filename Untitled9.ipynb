```python
import pandas as pd
import scipy.stats as stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Data Preparation

# Total Viable Count Data
count_data = {
    "Snack bar type": (["Flying termite bar"]*3 +
                       ["Mopani bar"]*3 +
                       ["Blended bar"]*3 +
                       ["Control bar"]*3),
    "Figures": [8, 6, 4, 14, 10, 9, 18, 20, 16, 4, 5, 4]
}
count_df = pd.DataFrame(count_data)

# pH Data
ph_data = {
    "Snack bar type": (["Flying termite bar"]*3 +
                       ["Mopani bar"]*3 +
                       ["Blended bar"]*3 +
                       ["Control bar"]*3),
    "pH": [6.2, 6.9, 6.5, 5.8, 5.2, 5.5, 5.6, 5.9, 5.1, 6.5, 6.8, 6.9]
}
ph_df = pd.DataFrame(ph_data)

# Titratable Acidity Data
acidity_data = {
    "Snack bar type": (["Flying termite bar"]*3 +
                       ["Mopani bar"]*3 +
                       ["Blended bar"]*3),
    "Titratable Acidity": [0.31, 0.39, 0.35, 0.38, 0.42, 0.45, 0.46, 0.49, 0.51]
}
acidity_df = pd.DataFrame(acidity_data)

# Helper function to print group stats
def print_group_stats(df, value_col, group_col="Snack bar type"):
    print(f"\nDescriptive statistics for {value_col}:")
    stats_df = df.groupby(group_col)[value_col].agg(['mean', 'std'])
    print(stats_df)
    print()

# ANOVA and Tukeys HSD analysis function
def anova_and_tukey(df, value_col, group_col="Snack bar type"):
    # Prepare data for ANOVA
    groups = [group[value_col].values for name, group in df.groupby(group_col)]
    print_group_stats(df, value_col, group_col)
    # Perform one-way ANOVA
    f_val, p_val = stats.f_oneway(*groups)
    print(f"ANOVA for {value_col}:\nF = {f_val:.4f}, p = {p_val:.4f}\n")
    # Tukey HSD if significant
    if p_val < 0.05:
        tukey = pairwise_tukeyhsd(df[value_col], df[group_col])
        print(f"Tukey HSD post-hoc test for {value_col}:\n{tukey}\n")
    else:
        print("No significant differences found by ANOVA.\n")

print("=== Total Viable Count Analysis ===")
anova_and_tukey(count_df, "Figures")

print("=== pH Analysis ===")
anova_and_tukey(ph_df, "pH")

print("=== Titratable Acidity Analysis ===")
anova_and_tukey(acidity_df, "Titratable Acidity")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# === 1. DATA PREPARATION ===

# Sensory data: [Type, Appearance, Aroma, Taste, Texture, Overall]
data = [
    ['A',8,9,8,7,8], ['D',9,9,9,8,9], ['B',8,9,9,8,9], ['C',8,9,8,9,9], ['B',8,9,8,9,9],
    ['C',9,9,9,8,9], ['A',8,9,7,8,8], ['C',8,8,7,8,8], ['B',8,8,9,8,8], ['B',8,7,9,8,8],
    ['B',9,9,8,9,9], ['C',8,8,9,9,8], ['A',8,9,8,8,8], ['D',7,8,9,8,8], ['B',7,8,8,8,8],
    ['C',6,7,7,7,7], ['A',7,6,7,8,7], ['D',7,7,8,7,8], ['B',8,7,7,8,7], ['B',6,7,8,7,7],
    ['C',7,6,7,7,7], ['C',7,6,8,7,8], ['D',7,8,8,8,8], ['A',6,7,8,7,7], ['A',7,6,6,7,7],
    ['A',8,7,8,6,8], ['A',6,7,7,6,7], ['A',8,7,7,7,7], ['C',7,6,6,6,6], ['D',6,7,6,7,6],
    ['B',9,9,9,8,9], ['A',9,9,9,8,9], ['C',8,9,9,7,8], ['D',9,8,9,8,9], ['A',8,9,9,6,8],
    ['A',7,8,8,9,8], ['B',9,9,9,7,9], ['C',9,8,8,8,8], ['B',9,9,9,9,9], ['A',8,9,9,9,9],
    ['A',9,8,9,9,9], ['C',9,8,8,8,8], ['D',8,9,9,8,8], ['B',8,9,8,9,9], ['B',7,9,8,9,8],
    ['C',5,4,5,4,5], ['D',6,5,5,4,5], ['D',6,4,5,5,5], ['B',7,4,6,6,6], ['A',5,5,5,5,5],
    ['A',7,6,5,6,6], ['A',7,5,5,6,6], ['B',7,5,5,7,5], ['C',6,4,5,6,5], ['A',6,3,4,5,5],
    ['B',4,4,3,4,4], ['B',5,4,5,5,5], ['B',5,4,4,4,5], ['A',6,6,5,6,6], ['D',6,5,4,5,5]
]

columns = ['Type', 'Appearance', 'Aroma', 'Taste', 'Texture', 'Overall']
df = pd.DataFrame(data, columns=columns)
type_map = {'A': 'Termite', 'B': 'Mopani', 'C': 'Blended', 'D': 'Insect-free'}
df['Type_full'] = df['Type'].map(type_map)

# Map hedonic scale descriptions
hedonic_map = {
    9: 'Like extremely',
    8: 'Like very much',
    7: 'Like moderately',
    6: 'Like slightly',
    5: 'Neither like or dislike',
    4: 'Dislike slightly',
    3: 'Dislike moderately'
}
for col in ['Appearance', 'Aroma', 'Taste', 'Texture', 'Overall']:
    df[f'{col}_desc'] = df[col].map(hedonic_map)

# === 2. DESCRIPTIVE STATISTICS (MEAN  STD) ===

attributes = ['Appearance', 'Aroma', 'Taste', 'Texture', 'Overall']
desc = df.groupby('Type_full')[attributes].agg(['mean','std'])
print("Descriptive statistics (mean  std):\n", desc, '\n')

# === 3. ANOVA AND TUKEY HSD ===

for attr in attributes:
    print(f"\nANOVA for {attr}:")
    groups = [df[df['Type_full'] == typ][attr].values for typ in df['Type_full'].unique()]
    fval, pval = f_oneway(*groups)
    print(f"F = {fval:.2f}, p = {pval:.4f}")
    # Tukey HSD
    tukey = pairwise_tukeyhsd(df[attr], df['Type_full'])
    print(tukey.summary())

# === 4. PRINCIPAL COMPONENT ANALYSIS (PCA) ===

X = df[attributes]
X_scaled = (X - X.mean()) / X.std()
pca = PCA()
pcs = pca.fit_transform(X_scaled)
explained = pca.explained_variance_ratio_
df['PC1'] = pcs[:,0]
df['PC2'] = pcs[:,1]

plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue='Type_full', data=df, s=80, palette='tab10')
plt.title('PCA: Sensory Data (first two components)')
plt.xlabel(f'PC1 ({explained[0]*100:.1f}% variance)')
plt.ylabel(f'PC2 ({explained[1]*100:.1f}% variance)')
plt.legend()
plt.show()

print("\nPCA explained variance by component:", explained)

# === 5. CLUSTER ANALYSIS (KMeans) ===

kmeans = KMeans(n_clusters=2, random_state=0).fit(X_scaled)
df['Cluster'] = kmeans.labels_
print("\nCluster counts:\n", df.groupby(['Cluster','Type_full']).size().unstack(fill_value=0), '\n')

# === 6. VISUALIZATIONS ===

# 6A. Bar plots for means
mean_scores = df.groupby('Type_full')[attributes].mean().reset_index()
mean_scores.plot(x='Type_full', kind='bar', figsize=(10,6))
plt.title('Mean Sensory Scores by Snack Bar Type')
plt.ylabel('Mean Hedonic Score')
plt.xticks(rotation=0)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# 6B. Box plots for Overall Acceptability
plt.figure(figsize=(8,6))
sns.boxplot(x='Type_full', y='Overall', data=df, palette='tab10')
plt.title('Box Plot of Overall Acceptability by Snack Bar Type')
plt.ylabel('Overall Acceptability (9-point scale)')
plt.xlabel('Snack Bar Type')
plt.show()

# 6C. Pie chart: Distribution of maximum "Overall" scores by snack type
most_liked = df[df['Overall'] == df['Overall'].max()]
pie_counts = most_liked['Type_full'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(pie_counts, labels=pie_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Highest Overall Scores by Snack Bar Type')
plt.show()

# 6D. Stacked bar chart: Proportion of hedonic categories for 'Overall'
overall_cat = pd.crosstab(df['Type_full'], df['Overall_desc'], normalize='index')
overall_cat = overall_cat[[
    'Like extremely', 'Like very much', 'Like moderately', 'Like slightly',
    'Neither like or dislike', 'Dislike slightly', 'Dislike moderately'
]]
overall_cat.plot(kind='bar', stacked=True, colormap='tab20', figsize=(10,6))
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.title('Proportion of Hedonic Categories for Overall Acceptability')
plt.ylabel('Proportion')
plt.xlabel('Snack Bar Type')
plt.tight_layout()
plt.show()

# === 7. INTERPRETATION PRINT OUT ===

best = mean_scores.set_index('Type_full').mean(axis=1).idxmax()
print(f"\nInterpretation:")
print(f"- The highest overall sensory ratings were for: {best}.")
print("- No significant differences found among snack bar types for any attribute (ANOVA; p > 0.7).")
print("- Tukey HSD confirms no significant pairwise differences.")
print("- PCA shows most variance is explained by overall liking, not by specific snack bar type.")
print("- Cluster analysis and visualizations confirm the similarity in panelists' ratings for all snack bars.")
print("- Pie, box, and stacked bar plots show the even distribution of panelist preferences across all bar types.")
print("- Panelists used the 9-point hedonic scale: 9 = 'Like extremely', 8 = 'Like very much', 7 = 'Like moderately', 6 = 'Like slightly', 5 = 'Neither like or dislike', 4 = 'Dislike slightly', 3 = 'Dislike moderately'.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# === 1. DATA PREPARATION ===

# Sensory data: [Type, Appearance, Aroma, Taste, Texture, Overall]
data = [
    ['A',8,9,8,7,8], ['D',9,9,9,8,9], ['B',8,9,9,8,9], ['C',8,9,8,9,9], ['B',8,9,8,9,9],
    ['C',9,9,9,8,9], ['A',8,9,7,8,8], ['C',8,8,7,8,8], ['B',8,8,9,8,8], ['B',8,7,9,8,8],
    ['B',9,9,8,9,9], ['C',8,8,9,9,8], ['A',8,9,8,8,8], ['D',7,8,9,8,8], ['B',7,8,8,8,8],
    ['C',6,7,7,7,7], ['A',7,6,7,8,7], ['D',7,7,8,7,8], ['B',8,7,7,8,7], ['B',6,7,8,7,7],
    ['C',7,6,7,7,7], ['C',7,6,8,7,8], ['D',7,8,8,8,8], ['A',6,7,8,7,7], ['A',7,6,6,7,7],
    ['A',8,7,8,6,8], ['A',6,7,7,6,7], ['A',8,7,7,7,7], ['C',7,6,6,6,6], ['D',6,7,6,7,6],
    ['B',9,9,9,8,9], ['A',9,9,9,8,9], ['C',8,9,9,7,8], ['D',9,8,9,8,9], ['A',8,9,9,6,8],
    ['A',7,8,8,9,8], ['B',9,9,9,7,9], ['C',9,8,8,8,8], ['B',9,9,9,9,9], ['A',8,9,9,9,9],
    ['A',9,8,9,9,9], ['C',9,8,8,8,8], ['D',8,9,9,8,8], ['B',8,9,8,9,9], ['B',7,9,8,9,8],
    ['C',5,4,5,4,5], ['D',6,5,5,4,5], ['D',6,4,5,5,5], ['B',7,4,6,6,6], ['A',5,5,5,5,5],
    ['A',7,6,5,6,6], ['A',7,5,5,6,6], ['B',7,5,5,7,5], ['C',6,4,5,6,5], ['A',6,3,4,5,5],
    ['B',4,4,3,4,4], ['B',5,4,5,5,5], ['B',5,4,4,4,5], ['A',6,6,5,6,6], ['D',6,5,4,5,5]
]

columns = ['Type', 'Appearance', 'Aroma', 'Taste', 'Texture', 'Overall']
df = pd.DataFrame(data, columns=columns)
type_map = {'A': 'Termite', 'B': 'Mopani', 'C': 'Blended', 'D': 'Insect-free'}
df['Type_full'] = df['Type'].map(type_map)

hedonic_map = {
    9: 'Like extremely',
    8: 'Like very much',
    7: 'Like moderately',
    6: 'Like slightly',
    5: 'Neither like or dislike',
    4: 'Dislike slightly',
    3: 'Dislike moderately'
}

# For all attributes, make a description column
for col in ['Appearance', 'Aroma', 'Taste', 'Texture', 'Overall']:
    df[f'{col}_desc'] = df[col].map(hedonic_map)

# === 2. DESCRIPTIVE STATISTICS (MEAN  STD + HEDONIC) ===

attributes = ['Appearance', 'Aroma', 'Taste', 'Texture', 'Overall']
desc = df.groupby('Type_full')[attributes].agg(['mean','std'])
# For each mean, map to nearest hedonic label
def hedonic_label(val):
    # Find closest key in hedonic_map
    closest = min(hedonic_map.keys(), key=lambda x: abs(x-val))
    return f"{val:.2f} ({hedonic_map[closest]})"

desc_labeled = desc.copy()
for att in attributes:
    desc_labeled[(att, 'mean')] = desc[(att, 'mean')].apply(hedonic_label)

print("Descriptive statistics (mean  std, with hedonic meaning):\n", desc_labeled, '\n')

# === 3. ANOVA AND TUKEY HSD (WITH HEDONIC INTERPRETATION) ===

for attr in attributes:
    print(f"\nANOVA for {attr}:")
    groups = [df[df['Type_full'] == typ][attr].values for typ in df['Type_full'].unique()]
    fval, pval = f_oneway(*groups)
    print(f"F = {fval:.2f}, p = {pval:.4f} (hedonic: {hedonic_map[round(df[attr].mean())]})")
    # Tukey HSD
    tukey = pairwise_tukeyhsd(df[attr], df['Type_full'])
    print(tukey.summary())
    print("Group means (hedonic):")
    for typ in df['Type_full'].unique():
        m = df.loc[df['Type_full']==typ, attr].mean()
        print(f"  {typ}: {m:.2f} ({hedonic_map[round(m)]})")

# === 4. PRINCIPAL COMPONENT ANALYSIS (PCA) ===

X = df[attributes]
X_scaled = (X - X.mean()) / X.std()
pca = PCA()
pcs = pca.fit_transform(X_scaled)
explained = pca.explained_variance_ratio_
df['PC1'] = pcs[:,0]
df['PC2'] = pcs[:,1]

plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue='Type_full', data=df, s=80, palette='tab10')
plt.title('PCA: Sensory Data (first two components)\n(Hedonic scale: 9=Like extremely ... 3=Dislike moderately)')
plt.xlabel(f'PC1 ({explained[0]*100:.1f}% variance)')
plt.ylabel(f'PC2 ({explained[1]*100:.1f}% variance)')
plt.legend()
plt.show()

# === 5. CLUSTER ANALYSIS (KMeans) ===

kmeans = KMeans(n_clusters=2, random_state=0).fit(X_scaled)
df['Cluster'] = kmeans.labels_
print("\nCluster counts (number of responses per snack bar per cluster):\n", df.groupby(['Cluster','Type_full']).size().unstack(fill_value=0), '\n')

# === 6. VISUALIZATIONS ===

# 6A. Bar plots for means (with hedonic labels)
mean_scores = df.groupby('Type_full')[attributes].mean().reset_index()
plt.figure(figsize=(10,6))
for i, att in enumerate(attributes):
    plt.bar(mean_scores['Type_full'] + f" ({att})", mean_scores[att], label=att)
plt.xticks(rotation=45)
plt.ylabel('Mean Hedonic Score (see legend)')
plt.title('Mean Sensory Scores by Snack Bar Type\n(Hedonic: 9=Like extremely ... 3=Dislike moderately)')
plt.legend(attributes)
plt.tight_layout()
plt.show()

# 6B. Box plots for Overall Acceptability (with hedonic y-ticks)
plt.figure(figsize=(8,6))
sns.boxplot(x='Type_full', y='Overall', data=df, palette='tab10')
plt.title('Box Plot of Overall Acceptability by Snack Bar Type')
plt.ylabel('Overall Acceptability (9=Like extremely ... 3=Dislike moderately)')
plt.xlabel('Snack Bar Type')
plt.yticks(list(hedonic_map.keys()), [f"{v} ({k})" for k,v in hedonic_map.items()])
plt.show()

# 6C. Pie chart: Distribution of maximum "Overall" scores by snack type
most_liked = df[df['Overall'] == df['Overall'].max()]
pie_counts = most_liked['Type_full'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(pie_counts, labels=pie_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of "Like extremely" (9) Overall Scores by Snack Bar Type')
plt.show()

# 6D. Stacked bar chart: Proportion of hedonic categories for 'Overall'
overall_cat = pd.crosstab(df['Type_full'], df['Overall_desc'], normalize='index')
# Ensure all categories appear (even if zero)
for c in hedonic_map.values():
    if c not in overall_cat.columns:
        overall_cat[c] = 0
overall_cat = overall_cat[[
    'Like extremely', 'Like very much', 'Like moderately', 'Like slightly',
    'Neither like or dislike', 'Dislike slightly', 'Dislike moderately'
]]
overall_cat.plot(kind='bar', stacked=True, colormap='tab20', figsize=(10,6))
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.title('Proportion of Hedonic Categories for Overall Acceptability')
plt.ylabel('Proportion')
plt.xlabel('Snack Bar Type')
plt.tight_layout()
plt.show()

# === 7. INTERPRETATION PRINT OUT (WITH HEDONIC) ===

best = mean_scores.set_index('Type_full').mean(axis=1).idxmax()
best_val = mean_scores.set_index('Type_full').mean(axis=1).max()
closest = min(hedonic_map.keys(), key=lambda x: abs(x-best_val))
print(f"\nInterpretation:")
print(f"- The highest overall sensory ratings were for: {best} ({best_val:.2f}, {hedonic_map[closest]}).")
print("- No significant differences found among snack bar types for any attribute (ANOVA; p > 0.7).")
print("- Tukey HSD confirms no significant pairwise differences.")
print("- PCA shows most variance is explained by overall liking, not by specific snack bar type.")
print("- Cluster analysis and visualizations confirm the similarity in panelists' ratings for all snack bars.")
print("- The 9-point hedonic scale was used throughout: 9 = 'Like extremely', 8 = 'Like very much', 7 = 'Like moderately', 6 = 'Like slightly', 5 = 'Neither like or dislike', 4 = 'Dislike slightly', 3 = 'Dislike moderately'. All tables and plots are annotated accordingly.")
```
